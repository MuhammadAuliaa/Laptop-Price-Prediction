```python
# import library yang diperlukan
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# load dataset dari file CSV
laptops_df = pd.read_csv("data/laptops_test.csv")
coba_df = pd.read_csv("data/laptops_train.csv")
laptops_df.head()
```

```python
# encoding kolom kategori menjadi angka
laptops_df["Manufacturer"] = laptops_df["Manufacturer"].astype('category').cat.codes
laptops_df["Model Name"] = laptops_df["Model Name"].astype('category').cat.codes
laptops_df["Category"] = laptops_df["Category"].astype('category').cat.codes
laptops_df["Screen"] = laptops_df["Screen"].astype('category').cat.codes
laptops_df["CPU"] = laptops_df["CPU"].astype('category').cat.codes
laptops_df["RAM"] = laptops_df["RAM"].astype('category').cat.codes
laptops_df["GPU"] = laptops_df["GPU"].astype('category').cat.codes
laptops_df["Operating System"] = laptops_df["Operating System"].astype('category').cat.codes
laptops_df["Operating System Version"] = laptops_df["Operating System Version"].astype('category').cat.codes
laptops_df["Weight"] = laptops_df["Weight"].astype('category').cat.codes
laptops_df[" Storage"] = laptops_df[" Storage"].astype('category').cat.codes

# memilih variabel independent dan dependent
X = laptops_df.drop("Price", axis=1)
X = laptops_df.drop("Screen Size", axis=1)
y = laptops_df["Price"]
```

```python
X
```

```python
# membagi dataset menjadi data training dan data testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# membuat model regresi linear dan mempelajari pola dari data training
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# melakukan prediksi harga laptop pada data testing
y_pred = regressor.predict(X_test)

# menampilkan hasil prediksi
for i in range(len(y_pred)):
    print("Harga laptop prediksi ke -", i+1, ": ", y_pred[i])

# menampilkan koefisien dari model regresi linear
print("Koefisien model regresi linear: ", regressor.coef_)
```
